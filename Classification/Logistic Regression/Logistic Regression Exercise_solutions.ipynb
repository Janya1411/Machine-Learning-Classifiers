#### convert categorical variables into numerical variables, make sure create a function for yes/ no columns

```python
df.head()
```

```python
df_job = pd.get_dummies(df['job'],drop_first=True)
df_marital = pd.get_dummies(df['marital'],drop_first=True)
df_education = pd.get_dummies(df['education'],drop_first=True)
df_contact = pd.get_dummies(df['contact'],drop_first=True)
df_month = pd.get_dummies(df['month'],drop_first=True)
df_poutcome = pd.get_dummies(df['poutcome'],drop_first=True)
```

```python
def change(x):
    if x == 'yes':
        return 1
    elif x == 'no':
        return 0
    else:
        return x
```

```python
df['default'] = df['default'].apply(change)
```

```python
df['housing'] = df['housing'].apply(change)
```

```python
df['loan'] = df['loan'].apply(change)
df['y'] = df['y'].apply(change)
```

```python
df.head()
```

#### concantenate all dataframes together

```python
df = pd.concat([df,df_job,df_marital,df_education,df_contact,df_month,df_poutcome],axis=1)
```

#### check the dataframe and see how many columns has been added

```python
df.head()
```

#### drop the categorical variables columns

```python
df.drop(['job','marital','education','contact','month','poutcome'],axis=1,inplace=True)
```

#### import train_test_split from sklearn

```python
from sklearn.cross_validation import train_test_split
```

#### Create variables X and y

```python
X = df.drop('y',axis=1)
y = df['y']
```

#### split the data into 80% training and 20% testing and random state as 101

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

#### Import LogisticRegression class from sklearn

```python
from sklearn.linear_model import LogisticRegression
```

#### create a object for logisticregression class

```python
lr = LogisticRegression()
```

#### fit the model into the object

```python
lr.fit(X_test,y_test)
```

#### predict the values and store it in a variable called y_pred

```python
y_pred = lr.predict(X_test)
```
